```python
# import 및 접속/로그인
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from urllib.parse import quote
import pandas as pd
import time
import random
import os
import matplotlib.pyplot as plt
from skimage import io
import requests

# 주피터 파일이랑 같은 폴더에 chromedriver 넣어야합니다.
driver = webdriver.Chrome('chromedriver.exe')
soup = bs(driver.page_source, 'html.parser')
driver.implicitly_wait(3)

basicUrl = "https://www.google.co.kr/imghp?hl=ko&tab=wi&ogbl"
driver.get(basicUrl)
keyword = input("수집하고 싶은 이미지를 입력해주세요 : ")
print("수집하고 싶은 페이지 수를 입력해주세요")
pages = input("해당 숫자만큼 스크롤을 내리게 됩니다 : ")

driver.find_element_by_name('q').send_keys(keyword)
driver.find_element_by_css_selector('svg').click()
```

```python
# hidden 페이지 해제 동작
from selenium.webdriver.common.keys import Keys
element=driver.find_element_by_tag_name('html')
for j in range(30*(int(pages)-1)):
        element.send_keys(Keys.ARROW_DOWN)
```

```python
import matplotlib.pyplot as plt
from skimage import io
import requests
image_blocks = driver.find_elements_by_css_selector('img.rg_ic')
image = [i.get_attribute('src') for i in image_blocks ]
print("Total image : " + str(len(image)))
print("===============================")
saveCount = 0
count = 0
for image_url in image:
    count += 1
    print("Count : " + str(count))

    try:
        #print(image_url)
        response = requests.get(image_url)
        imageData = response.content
        filename = "image/" + keyword + "/" + keyword + str(count) + ".jpg"

        with open(filename, 'wb') as f:
            print("File name : " + filename + " Save Completed")
            f.write(imageData)
        saveCount += 1
    except:
        print("error detected")
    
    print("Save Count : " + str(saveCount))    
    print("=======================")
print("Save Complete")
print("Save Count : " + str(saveCount))
```
